```
import gzip
import pandas as pd
```

```
db=pd.read_json('http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz',lines=True)
db
```

```

db['reviewText'] = db['reviewText'].str.lower()#makes characters lower-cased
db['summary'] = db['summary'].str.lower()
```

```

def count_words(text):
  word_list = text.split() #returns list of strings with characters as elements
  total_words = len(text.split())# total words in the string
  # Initialize Dictionary
  d = {'total_count' : total_words}# counts number of times each word comes up in dictionary
  for w in word_list: 
     d[w] = d.get(w, 0) + 1  # value gets incremented if the word occurs again
  return d
```

```
db['reviewText_wordcount'] = db.apply(lambda row : count_words(row['reviewText']), axis = 1) # apply() function in order to apply count_words function to every row of the column
db['summary_wordcount'] = db.apply(lambda row : count_words(row['summary']), axis = 1)
```
